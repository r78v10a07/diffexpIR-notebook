## Getting started

Follow this gide to install the basic software and data requiered by this project

## Notebook

This is a Jupyter Notebook to show how to install all software required by this project

### Requirements
#### Basic programs
1. Python3
2. virtualenv3
3. C++

#### Bioinformatics programs
These programs have to be installed and included in the user PATH variable.

* STAR (https://github.com/alexdobin/STAR)
* SAMTools (http://www.htslib.org/)
* BAMTools (https://github.com/pezmaster31/bamtools)
* TPMCalculator (https://github.com/r78v10a07/TPMCalculator)
* DiffExpIR (https://github.com/r78v10a07/DiffExpIR)

### Instalation of the Notebook

Execute this commands in a BASH terminal just the first time to get the notebook dependencies installed.

```
$ git clone https://github.com/r78v10a07/diffexpir_data
$ cd diffexpir_data

# Creating a virtual environment
$ virtualenv3 venv
$ source venv/bin/activate
$ pip install -r requirements/jupyter.txt
$ pwd
```

Edit the file in config/proj_env and change the PROJ_ROOT variable (first line)

### Downloading data files

Download this file ftp://ftp.ncbi.nlm.nih.gov/pub/marino/internal/other/roberto/diffexpIR-notebook-data.tar.gz from our public FTP.

Uncompress this file to create the data folder.

```
$ wget ftp://ftp.ncbi.nlm.nih.gov/pub/marino/internal/other/roberto/diffexpIR-notebook-data.tar.gz
$ tar xzfv diffexpIR-notebook-data.tar.gz
```

### Instalation of software 

#### Compiling the STAR

```
$ cd src/
$ git clone https://github.com/alexdobin/STAR
$ cd STAR/source
$ make STAR
$ cd ../../../bin/
$ ln -s ../src/STAR/source/STAR STAR
$ cd ../
$ cd data/hg19/
$ mkdir STAR
$ cd STAR
$ ln -s ../genome.fa genome.fa
$ ln -s ../genes.gtf genes.gtf
$ ../../../bin/STAR --runThreadN 16 --runMode genomeGenerate --genomeDir . --genomeFastaFiles genome.fa --sjdbGTFfile ./genes.gtf
$ cd ../../../
```

##### Configuring start.sh script in bin

Edit the script bin/star.sh

Modify ROW 7 setting the correct path to the STAR software. Then, modify the ROW 8 to set the correct path to the samtools.

#### Compiling the BAMTools

```
$ cd src/
$ git clone https://github.com/pezmaster31/bamtools
$ cd bamtools
$ mkdir build
$ cd build
$ cmake ..
$ make
$ cd ../../../bin/
$ ln -s ../src/bamtools/bin/bamtools bamtools
$ cd ../
$ source config/proj_env
```

#### Compiling TPMCalculator

```
$ cd src/
$ git clone https://github.com/r78v10a07/TPMCalculator
$ cd TPMCalculator
$ make 
$ cd ../../bin/
$ ln -s ../src/TPMCalculator/bin/TPMCalculator TPMCalculator
$ cd ../
```

#### Compiling DiffExpIR

```
$ cd src/
$ git clone https://github.com/r78v10a07/DiffExpIR
$ cd DiffExpIR
$ make 
$ cd ../../bin/
$ ln -s ../src/DiffExpIR/bin/diffexpIR diffexpIR
$ cd ../
```
